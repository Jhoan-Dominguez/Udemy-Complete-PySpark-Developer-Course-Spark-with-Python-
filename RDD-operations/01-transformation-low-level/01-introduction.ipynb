```python
import findspark
findspark.init('/home/jhoand/spark-3.2.1-bin-hadoop3.2/')
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('test').getOrCreate()
sc = spark.sparkContext

def print_function(rdd, num = 10):
    for i in rdd.take(num): print(i)
    print('\n')
```

```python
rdd = sc.textFile('/home/jhoand/course/RDD-operations/text-fields/PracticeFiles/Orders')
print_function(rdd)
```

```python
#rdd from list
data = [1,2,3,4,5,6,7]
rdd = spark.sparkContext.parallelize(data)

for i in rdd.take(5): print(i)
```

```python
new_rdd = rdd.map(lambda x: x * 2)
for i in new_rdd.take(5): print(i)
```
